** Recreate the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.**

** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

g= sns.jointplot(x='fare', y='age', data=titanic, kind="scatter")

# Add labels and a title
g.set_axis_labels("Fare", "Age")
g.fig.suptitle("Jointplot with Scatter Plot and Marginal Histplots", y=1.02)

# Show the plot
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

sns.distplot(titanic["fare"], kde=False)

# Add labels and a title
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Distribution Plot for Fare')

# Show the plot
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!


plt.figure(figsize=(10, 6))  
sns.boxplot(x='pclass', y='age', data=titanic)

plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Boxplot of Age by Passenger Class')

plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
plt.figure(figsize=(5, 4))  # Adjust the size as needed

# Create a swarmplot with different colors for each passenger class
sns.swarmplot(x='pclass', y='age', data=titanic, hue='class', palette='Set1')

# Add labels and a title
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Swarmplot of Age by Passenger Class')

# Show the legend
plt.legend(title='Class')

# Show the plot
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

plt.figure(figsize=(4, 3))  

sns.countplot(x='sex', data=titanic)

plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Count Plot of Passengers by Sex')

plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!


tc = titanic.corr()

plt.figure(figsize=(5, 4))  # Adjust the size as needed

sns.heatmap(tc, cmap='coolwarm', annot=True, fmt=".2f")

plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

g = sns.FacetGrid(data=titanic, col='sex')

g.map(sns.histplot, 'age', kde=False)

g.set_axis_labels('Age')
g.set_titles(col_template='{col_name}')

plt.tight_layout()

plt.show()
```
